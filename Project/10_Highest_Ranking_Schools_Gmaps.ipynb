```python
import gmaps
import numpy as np
import pandas as pd
import requests
import time
import json
from pprint import pprint
from config import gkey
```

```python
highest_schools = pd.read_csv("output_data/highest_schools.csv")
highest_schools = highest_schools.drop(columns="Unnamed: 0")
highest_schools
```

```python
gmaps.configure(api_key=gkey)
locations = highest_schools[["HOME_LAT", "HOME_LNG"]]
avg_home_prices = highest_schools["AVG HOME PRICES"]
```

```python
fig = gmaps.figure()
heat_layer = gmaps.heatmap_layer(locations, weights=avg_home_prices,
                                 dissipating=False, max_intensity=100,
                                 point_radius = 5)
fig.add_layer(heat_layer)
fig
```

```python
marker_locations = highest_schools[["SCHOOL_LAT", "SCHOOL_LNG"]]

fig = gmaps.figure()
markers = gmaps.marker_layer(marker_locations)
fig.add_layer(markers)
fig
```
