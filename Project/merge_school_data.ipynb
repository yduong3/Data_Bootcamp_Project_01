# Merging School Data

```python
import pandas as pd

# Output File (CSV)
output_data_tcdsb = "output_data/toronto_catholic_schools.csv"
output_data_tdsb = "output_data/toronto_schools.csv"
```

```python
# Take in all of our school data and read it into pandas
tcdsb_schools = "Resources/Schools_TCDSB.csv"
tcdsb_ratings = "Resources/Schools_TCDSB_Ratings.csv"
tdsb_schools = "Resources/Schools_TDSB.csv"
tdsb_ratings = "Resources/Schools_TDSB_Ratings.csv"


tcdsb_schools_df = pd.read_csv(tcdsb_schools, encoding='latin-1')
tcdsb_ratings_df = pd.read_csv(tcdsb_ratings, encoding='latin-1')
tdsb_schools_df = pd.read_csv(tdsb_schools, encoding='latin-1')
tdsb_ratings_df = pd.read_csv(tdsb_ratings, encoding='latin-1')
```

```python
# Merge Toronto Catholic School Board datasets on "School Name" so that no data is lost
combined_tcdsb_df = pd.merge(tcdsb_schools_df, tcdsb_ratings_df, how='outer', on='SCHOOL NAME')
combined_tcdsb_df.head(20)
```

```python
# Dropping NaN and changing index to School Name
combined_tcdsb_df.dropna(how="any", inplace=True)
combined_tcdsb_df.set_index("SCHOOL NAME", inplace = True)
combined_tcdsb_df.head(20)
```

```python
# Export the Toronto Catholic School Board data into a csv
combined_tcdsb_df.to_csv(output_data_tcdsb, index_label="SCHOOL NAME")
```

```python
# Merge Toronto School Board datasets on "School Name" so that no data is lost
combined_tdsb_df = pd.merge(tdsb_schools_df, tdsb_ratings_df, how='outer', on='SCHOOL NAME')
combined_tdsb_df.head(20)
```

```python
# Dropping NaN and changing index to School Name
combined_tdsb_df.dropna(how="any", inplace=True)
combined_tdsb_df.set_index("SCHOOL NAME", inplace = True)
combined_tdsb_df.head(20)
```

```python
# Export the Toronto School Board data into a csv
combined_tdsb_df.to_csv(output_data_tdsb, index_label="SCHOOL NAME")
```
