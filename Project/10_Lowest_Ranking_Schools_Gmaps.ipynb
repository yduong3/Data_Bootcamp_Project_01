```python
import gmaps
import numpy as np
import pandas as pd
import requests
import time
import json
from pprint import pprint
from config import gkey
```

```python
lowest_schools = pd.read_csv("output_data/lowest_schools_yd.csv")
lowest_schools = lowest_schools.drop(columns="Unnamed: 0")
lowest_schools
```

```python
gmaps.configure(api_key=gkey)
locations = lowest_schools[["HOME_LAT", "HOME_LNG"]]
avg_home_prices = lowest_schools["WEIGHTS"]
```

```python
fig = gmaps.figure()
heat_layer = gmaps.heatmap_layer(locations, weights=avg_home_prices,
                                 dissipating=False, max_intensity=100,
                                 point_radius = 5)
fig.add_layer(heat_layer)
fig
```

```python
marker_locations = lowest_schools[["SCHOOL_LAT", "SCHOOL_LNG"]]

fig = gmaps.figure()
markers = gmaps.marker_layer(marker_locations)
fig.add_layer(markers)
fig
```
